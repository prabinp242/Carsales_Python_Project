```python
df['Total Sales'] = df[['Q1 Sales','Q2 Sales','Q3 Sales','Q4 Sales']].sum(axis=1)
```

```python
# How can you find the brand with the highest total sale

# idxmax = index maximum value
# idxmin = index minimum value

df.loc[df['Total Sales'].idxmax()].to_frame()
```

```python
# How can you add a new column that calculates the average quarterly sales for each brand

df['Avg Quarterly Sales'] = df[['Q1 Sales', 'Q2 Sales', 'Q3 Sales', 'Q4 Sales']].mean(axis=1)
```

```python

# How can you plot the total sales for each brand using a bar chart

import matplotlib.pyplot as plt

df.groupby('BRAND')['Total Sales'].sum().plot(kind='barh')
plt.title('Total Sales per Brand')
plt.ylabel('Total Sales')
plt.show()
```

```python
# How can you calculate the percentage contribution of each quarter to the total sales for each brand
# Formula = (Q1 + Q2 + Q3 + Q4) / Total Sales * 100

quarters = ['Q1 Sales', 'Q2 Sales', 'Q3 Sales', 'Q4 Sales']

for i in quarters:
    df[f"{i} %"] = (df[i] / df['Total Sales']) * 100

print(df[[i for i in quarters] + [f"{i} %" for i in quarters]])
```

```python
quarters = ['Q1 Sales', 'Q2 Sales', 'Q3 Sales', 'Q4 Sales']

# All Quarters - Total Sales
# Formula = (Q1 + Q2 + Q3 + Q4) / Total Sales * 100

df[[f"{i} %" for i in quarters]] = df[quarters].apply(lambda x: (x / df['Total Sales']) * 100)
print(df)
```

```python
# How can you create a pivot table to show the total sales for each brand per quarter

df.pivot_table(index='BRAND',
               values=quarters,
               aggfunc='sum')
```

```python
# How can you rename the columns Q1 Sales, Q2 Sales, Q3 Sales, and Q4 Sales to Quarter 1, Quarter 2, Quarter 3, Quarter 4

df.rename(columns={
    'Q1 Sales': 'Quarter 1',
    'Q2 Sales': 'Quarter 2',
    'Q3 Sales': 'Quarter 3',
    'Q4 Sales': 'Quarter 4'
}, inplace=True)
```

```python
# How can you calculate the year-over-year sales growth for each brand

df['YoY Growth (%)'] = ((df['Total Sales'] - df['Last Year Sales']) / df['Last Year Sales']) * 100
```
